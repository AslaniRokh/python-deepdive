# Multi-Line Statements and Strings
### Certain physical newlines are ignored in order to form a complete logical line of code.



### Implicit Examples


```python
a = [1, 2, 3]
```

```python
b = (1, 2, 3)
```

```python
c = {'key1': 1, 'key2': 2, 'key3': 3}
```

```python
def my_func(a, b, c):
    print(a, b, c)
```

```python
a = [1, 2, 3,
    4, 5, 6]
```

```python
b = (1, 2, 3,
    4, 5, 6)
```

```python
c = {'key2': 1, 'key2': 2, 'key3': 3,
    'key4': 4, 'key5': 5, 'key6': 6}
```

```python
def my_func(a, b, c,
           d, e, f):
    print(a, b, c, d, e, f)
```

```python
a
```

```python
b
```

```python
c
```

```python
my_func(1, 2, 3,
       4, 5, 6)
```

#### You may also add comments to the end of each physical line:



```python
a = [1, # item 1
    2, # item 2
    3, # item 3
    ]
```

```python
b = (1, # item 1
    2, # item 2
    3, # item 3
    )
```

```python
c = {'key1': 1, # value for key 1
  'key2': 2, # value for key 2
  'key3': 3, # value for key 3
  }
```

```python
def my_func(a, #comment
           b, #comment
           c, #comment
           ):
    print(a, b, c)
```

```python
a
```

```python
b
```

```python
c
```

```python
my_func(1, 2, 3)
my_func(1, #comment
       2, #comment
       3, #comment
       )
```

#### other way of adding comments to the end of each physical line:

```python
a = [1 # item 1
    ,2 # item 2
    ,3 # item 3
    ]
```

```python
b = (1 # item 1
    ,2 # item 2
    ,3 # item 3
    )
```

```python
c = {'key1': 1 # value for key 1
    ,'key2': 2 # value for key 2
    ,'key3': 3 # value for key 3
    }
```

```python
def my_func(a # comment
            ,b # comment
            ,c # comment
           ):
    print(a, b, c)
```

```python
a
```

```python
b
```

```python
c
```

```python
my_func(1, 2, 3)
my_func(1 # comment
       ,2 #comment
       ,3 #comment
       )
```

###### Note if you do use comments, you must close off the collection on a new line.

###### i.e. the following will not work since the closing ] is actually part of the comment:

```python
a = [1 # comment ,
    2 # comment , 
    3 # comment]
```

```python
a = [1, # comment
    2, # comment
    3, # comment]
```

### Explicit Examples
#### You can use the \ character to explicitly create multi-line statements.




```python
a = 10
b = 20
c = 30
if a > 5 and b > 10 and c > 20:
    print('yes')
```

```python
if a > 5 \
    and b > 10 \
    and c > 20:
    print('yes')
```

##### The identation in continued-lines does not matter:

```python
if a > 5 \
and b > 10 \
and c > 20:
    print('yes')
```

### Multi-Line Strings

```python
a
```

```python
a = '''this
is a string'''
```

```python
a
```

###### Be careful if you indent your multi-line strings - the extra spaces are preserved!



```python


a = '''this
    is a string
        that is created over multiple lines'''
```

```python
print(a)
```

```python
def my_func():
    a = '''a multi-line string
    that is actually indented in the second line'''
    return a
```

```python
print(my_func())
```

```python
def my_func():
    a = '''a multi-line string
that is not indented in the second line'''
    return a
```

```python
print(my_func())
```
