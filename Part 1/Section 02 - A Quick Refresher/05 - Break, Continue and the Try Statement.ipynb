 ## Loop Break and Continue inside a Try...Except...Finally
Recall that in a <code>try</code> statement, the <code>finally</code> clause always runs:



```python
# try exception finally
a = 10
b = 0
try:
    a / b
except ZeroDivisionError:
    print('Division by zero')
finally:
    print('this always executes')
```

So, what happens when using a <code>try</code> statement within a <code>while</code> loop, and a <code>continue</code> or <code>break</code> statement is encountered?

```python
a = 0
b = 2

while a < 3:
    print('-----------------------------')
    
    a += 1
    b -= 1
    
    try:
        a / b
        
    except ZeroDivisionError:
        print('{0}, {1} - division by 0'.format(a, b))
        continue
    
    finally:
        print('{0}, {1} - always executes'.format(a, b))
        
    print('{0}, {1} - main loop'.format(a, b))
```

```python
a = 0
b = 2

while a < 3:
    print('-----------------------------')
    
    a += 1
    b -= 1
    
    try:
        a / b
        
    except ZeroDivisionError:
        print('{0}, {1} - division by 0'.format(a, b))
        break
    
    finally:
        print('{0}, {1} - always executes'.format(a, b))
        
    print('{0}, {1} - main loop'.format(a, b))
```

combine with else clause

We can even combine all this with the <code>else</code> clause:



```python
a = 0
b = 10

while a < 3:
    print('-----------------------------')
    
    a += 1
    b -= 1
    
    try:
        a / b
        
    except ZeroDivisionError:
        print('{0}, {1} - division by 0'.format(a, b))
        break
    
    finally:
        print('{0}, {1} - always executes'.format(a, b))
        
    print('{0}, {1} - main loop'.format(a, b))

else:
    print('Code executed without a zero division error')
```
