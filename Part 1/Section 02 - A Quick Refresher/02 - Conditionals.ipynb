### Conditionals
#### A conditional is a construct that allows you to branch your code based on conditions being met (or not)
<hr>

#### This is achieved using <b>if, elif</b> and <b> else</b> or the ternary operator (aka conditional expression)

```python
a = 2

if a < 5:
    print('a < 5')
```

```python
a = 6

if a < 5:
    print('a < 5')

else:
    print('a >= 5 ')
```

#### <b> if </b>statements can be nested:



```python
a = 5
if a < 5:
    print('a < 5')

else:
    if a < 10:
        print('5 <= a < 10')
        
    else:
        print('a >= 10')
```

#### But the elif statement provides far better readability:

```python
a = 15

if a < 5:
    print('a < 5')
    
elif a < 10:
    print('5 <= a < 10')

elif a < 15:
    print('10 <= a < 15')

elif a < 20:
    print('15 <= a < 20')
    
else:
    print('a > 20')
```

###### In Python, <u>elif</u> is the closest you'll find to the <u>switch / case</u> statement available in some other languages.

#### Python also provides a conditional expression (ternary operator):

### X if (condition) else Y

```python
a = 25

if a < 15:
    b = 'a < 5'

else:
    b = 'a >= 5'
    
print(b)
```

```python
a = 5
res = 'a < 10' if a < 10 else 'a >= 10'
print(res)
```

```python
a = 15
res = 'a < 15' if a < 15 else 'a >= 15'
print(res)
```

##### Note that X and Y can be any expression, not just literal values:



```python
def say_hello():
    print('Hello!')

def say_goodby():
    print('Goodbye!')
```

```python
a = 5
say_hello() if a < 6 else say_goodby()
```

```python
a = 15
say_hello() if a > 20 else say_goodby()
```
